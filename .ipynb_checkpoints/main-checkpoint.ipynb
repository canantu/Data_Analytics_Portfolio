```python
import pandas as pd
import matplotlib.pyplot as plt
import re
import string
from wordcloud import WordCloud 
import spacy
from nltk.corpus import stopwords

from sklearn.feature_extraction.text import CountVectorizer
from collections import  Counter

pd.options.display.max_columns = 999
```

```python
data= pd.read_csv('result_BFord.csv')
data.head()
```

```python
df = data.copy()
```

```python
def clean_text(text):
    
    #remove white spaces
    text = text.replace('\\n', ' ').replace('\n', ' ').replace('\t',' ').replace('\\', ' ').replace('. com', '.com')
    
    #remove http, https, www
    text = re.sub(r'https?://\S+|www\.\S+', r'',text) 
    
    #remove .com
    text = re.sub(r'[/A-Za-z]*\.com[/A-Za-z]*', r'', text) 
    
    #remove @words
    text= re.sub(r'(\s)@\w+|@\w+',r'',text)
    
    #remove stopwords
    sw = stopwords.words('english')
    for i in ["i'm",'i’m', "i'd", "i'll", "there's", "there're", "he's", "he'd", "he'll", "she's", "she'd", "she'll", 
          "they're", "they'd", "they'll", "they've", "we're", "we've", "we'll", "it'll", "it'd", 'us', 'via', 'ive', 'weve', 'theyve', "you'll", 'youll']:
        sw.append(i)
        
    text = [word.lower() for word in text.split() if word.lower() not in sw]
    text= " ".join(text)
    
    
    #remove punctuation
    punc = '!"#$%&\'()*+,-./:;<=>?@[\\]^_`{|}~´’<>°≤‘±–“”'
    translator = str.maketrans('', '', punc)
    text= text.translate(translator)
    
    #remove rt, tbt, rts, pt, gmt, pm, am
    text = re.sub(r'pm\s|pm$|am\s|am$|tbt\s|tbt$|rt\s|rt$|rts\s|rts$|gmt\s|gmt$|pt\s|pt$|', r'', text)
    
    #remove numbers
    text = re.sub(r'\w*\d\w*', r'', text)
    
    #remove white spaces
    text = text.replace('\\n', ' ').replace('\n', ' ').replace('\t',' ').replace('\\', ' ').replace('. com', '.com')
    
    return text
    
```

```python
def df_lemma(df):
    #lemmatization
    nlp = spacy.load('en')
    doc = df['tweet_text'].apply(nlp)
    
    for i, sent in enumerate(doc):
        df.loc[i, 'lemma']=' '.join([token.lemma_ for token in sent])
        
    df['lemma']=df['lemma'].apply(lambda text: re.sub('-PRON-',r'',text))
    
    return df
```

```python
def display_wordcloud(df):

    corpus_lemma=[]
    for sent in df['lemma']:
        for word in sent.split():
            corpus_lemma.append(word)
    
    wordcloud = WordCloud(width=800, height=400).generate(str(corpus_lemma))

    plt.figure(figsize=(20,10))
    plt.imshow(wordcloud)
    plt.axis("off")
    plt.show()
```

```python
def make_bow(df):
    
    #bag of words
    vectorizer = CountVectorizer(analyzer = 'word', lowercase=False)
    X = vectorizer.fit_transform(df["lemma"])
    bow = pd.DataFrame(X.toarray(), columns=vectorizer.get_feature_names())
    df_bow = pd.concat([bow, df[["lemma"]]], axis=1)

    return df_bow
```

```python
df['tweet_text']=df['tweet_text'].apply(lambda x: clean_text(x))
df_lemma(df)
df_bow = make_bow(df)
```

```python
df_bow.to_csv('df_bow.csv', index=False)
```

```python
display_wordcloud(df)
```
